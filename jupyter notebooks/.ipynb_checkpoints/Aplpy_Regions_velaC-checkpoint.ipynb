```python
"""Example how to plot sky regions on a sky image.
"""
from astropy.coordinates import SkyCoord, Angle
from regions import (
    make_example_dataset,
    CircleSkyRegion,
)
import matplotlib.pyplot as plt

config = dict(crpix=(18, 9), cdelt=(-10, 10), shape=(18, 36))

# generate pseudo data here
dataset = make_example_dataset(data='simulated', config=config) 
wcs = dataset.wcs

fig = plt.figure()
ax = fig.add_axes([0.15, 0.1, 0.8, 0.8], projection=wcs)

ax.set_xlim(-0.5, dataset.config['shape'][1] - 0.5)
ax.set_ylim(-0.5, dataset.config['shape'][0] - 0.5)

#dataset.image.data
ax.imshow(psi_table, cmap='gray', vmin=0, vmax=1,
          interpolation='nearest', origin='lower')

for source in dataset.source_table:
    # Plot a sky circle around each source
    center = SkyCoord(source['GLON'], source['GLAT'], unit='deg', frame='galactic')
    radius = Angle(20, 'deg')
    region = CircleSkyRegion(center=center, radius=radius)
    pix_region = region.to_pixel(wcs=wcs)
    pix_region.plot(ax=ax, edgecolor='yellow', facecolor='yellow', alpha=0.5, lw=3)
    print()
```

```python
# circular color maps
# deep masks
# Use masking to get rid of the edges! (deep mask file)

deep_mask = fits.open("VelaC_deepmap_mask.fits")
masking = deep_mask[0].data

plt.figure()
plt.imshow(masking)
plt.colorbar()

for i in range(masking.shape[0]):
    for j in range(masking.shape[1]):
        if masking[i][j] == 1:
            masking[i][j] = 0
        elif masking[i][j] == 0:
            masking[i][j] = 1

plt.figure()
plt.imshow(masking)
plt.colorbar()
```

```python
# Now show the masking version
plt.figure(figsize = (16, 12))
plt.title("James Calculated Psi Map")
plt.imshow(psi_table * masking, origin='lower')
plt.colorbar()
```

```python
"""
Let's take a look at the actual psi map
"""
from astropy.wcs import WCS

real_psi = fits.open("VelaC_500_intermediate_regrid_30as_pix_ang.fits")
real_int = fits.open("VelaC_500_intermediate_regrid_30as_pix_I.fits")[0].data
real_psi_data = real_psi[0].data
header = real_psi[0].header

plt.figure(figsize = (16, 12))
plt.title("Laura Calculated Psi Map")
plt.imshow(real_psi_data, origin = 'lower')
plt.colorbar()
plt.show()

wcs_helix = WCS(header)
fig = plt.figure(figsize=(10, 10))
ax = plt.subplot(projection=wcs_helix)
plt.imshow(real_int, aspect='equal')
plt.contour(real_int, 10, cmap='RdGy')
plt.xlabel(r'RA')
plt.ylabel(r'Dec')

# overlay = ax.get_coords_overlay('icrs')
# overlay.grid(color='white', ls='dotted')
```
