```python
def autocorr(x):
    result = np.correlate(x, x, mode='full')
    return result[result.size/2:]

def Imshow(image, **kwargs):
    """
    Simple function to an image.
    """
    plt.figure(figsize=(8, 6))
    plt.imshow(image)
    plt.colorbar()
    plt.show()
```
